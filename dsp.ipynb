```python
# Step 2: Import Libraries
import sys
import numpy as np
from funasr import AutoModel
from google.colab import files
from IPython.display import Audio, display
import base64
import io

print("✅ Libraries imported successfully!")
```

```python
# Step 3: Load the Pre-trained Model
print("\n🤖 Loading emotion detection model...")
print("⏳ This will take 1-2 minutes on first run...")

try:
    model = AutoModel(model="emotion2vec_plus_large")
    print("✅ Model loaded successfully!")
except Exception as e:
    print(f"❌ Error loading model: {e}")
    sys.exit(1)

print("\n🎭 EMOTION DETECTION READY!")
print("=" * 50)
```

```python
# Function 1: Upload and Analyze Audio File
def analyze_uploaded_audio():
    """Upload and analyze an audio file"""
    print("\n📁 UPLOAD AUDIO FILE")
    print("Please upload a .wav audio file:")

    try:
        uploaded = files.upload()

        if not uploaded:
            print("❌ No file uploaded")
            return

        filename = list(uploaded.keys())[0]
        print(f"📄 Uploaded: {filename}")

        # Analyze emotion
        print("🧠 Analyzing emotion...")
        result = model.generate(filename, granularity="utterance")

        # Display result - Handle the actual format
        if result and len(result) > 0:
            emotion_data = result[0]  # Get first result

            # Extract emotion labels and scores
            labels = emotion_data['labels']
            scores = emotion_data['scores']

            # Find the emotion with highest score
            max_score_idx = scores.index(max(scores))
            detected_emotion = labels[max_score_idx]
            confidence = scores[max_score_idx]

            # Clean up the emotion label (remove Chinese part)
            if '/' in detected_emotion:
                emotion_clean = detected_emotion.split('/')[-1]  # Get English part
            else:
                emotion_clean = detected_emotion

            # Display results
            print(f"\n🎭 DETECTED EMOTION: {emotion_clean.upper()}")
            print(f"📊 Confidence: {confidence:.2%}")

            # Show top 3 emotions
            print(f"\n📈 TOP 3 EMOTIONS:")
            sorted_emotions = sorted(zip(labels, scores), key=lambda x: x[1], reverse=True)
            for i, (emotion, score) in enumerate(sorted_emotions[:3]):
                emotion_name = emotion.split('/')[-1] if '/' in emotion else emotion
                print(f"   {i+1}. {emotion_name}: {score:.2%}")

        else:
            print("❌ No emotion detected - empty result")

    except Exception as e:
        print(f"❌ Error analyzing audio: {e}")
```

```python
analyze_uploaded_audio()
```
